```python
import numpy as np
import pandas as pd
import tensorflow as tf
import math
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
x = []
y = []

x_test = []

cabins = {'C': 0, 'Q': 1, 'S': 2}

for i in range(len(train)):
    x.append([
        train['Pclass'][i],
        1 if train['Sex'][i] == 'male' else 0,
        train['Age'][i] if train['Age'][i] > 0 else -50,
        train['SibSp'][i],
        train['Parch'][i],
        train['Fare'][i],
        cabins[train['Embarked'][i]] if train['Embarked'][i] in ('C', 'Q', 'S') else -50,
    ])
    
    y.append(train['Survived'][i])

for i in range(len(test)):
    x_test.append([
        test['Pclass'][i],
        1 if test['Sex'][i] == 'male' else 0,
        test['Age'][i] if test['Age'][i] > 0 else -50,
        test['SibSp'][i],
        test['Parch'][i],
        test['Fare'][i],
        cabins[test['Embarked'][i]] if test['Embarked'][i] in ('C', 'Q', 'S') else -50,
    ])

x = np.array(x)
y = np.array(y)

x_test = np.array(x_test)

print(x)
print(y)
print(x_test)
```

```python
l = tf.keras.layers

model = tf.keras.models.Sequential([
    l.Dense(128, input_shape=(7,), activation='relu'),
    l.Dropout(0.5),
    l.Dense(32),
    l.Dropout(0.5),
    l.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

model.summary()
```

```python
model.fit(x, y, epochs=100, batch_size=10, verbose=1)
```

```python
predictions = model.predict(x)

for i in range(len(train)):
    print(train['Name'][i])
    print(round(predictions[i][0]*100, 2), '% de chance que esteja vivo(a) hoje.')
    print('E ele(a) está.' if train['Survived'][i] == 1 else 'E ele(a) não está.')
    print()
```

```python
predictions = model.predict(x_test)

for i in range(len(test)):
    print(test['Name'][i])
    print(round(predictions[i][0]*100, 2), '% de chance que esteja vivo(a) hoje.')
    # print('E ele(a) está.' if train['Survived'][i] == 1 else 'E ele(a) não está.')
    print()
```

```python
submission = {
    'PassengerId': [],
    'Survived': []
}

predictions = model.predict(x_test)

test.head()
for i in range(len(test)):
    passengerId = test['PassengerId'][i]
    survived = 1 if predictions[i][0] > 0.5 else 0
    
    submission['PassengerId'].append(passengerId)
    submission['Survived'].append(survived)

submission_csv = pd.DataFrame(submission)

submission_csv.head()
submission_csv.reset_index(drop=True, inplace=True)

submission_csv.to_csv('submission.csv')
```
